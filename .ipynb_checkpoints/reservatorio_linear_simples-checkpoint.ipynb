```python
import numpy as np
import hydroeval as he
from scipy import integrate

import matplotlib.pyplot as plt
# Ajustando gráficos inline
plt.rcParams['figure.dpi'] = 150
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
```

Ajuste o modelo Reservatório Linear Simples aos dados apresentados na tabela a seguir.
Considerar os dados como sendo de precipitação efetiva e escoamento superficial.
Estabeleça $K$ de diferentes formas e verifique a adequação do ajuste.

| Tempo | Precipitação (mm) | Vazão (mm/$\Delta$t) |
|-------|-------------------|----------------------|
| 1     | 1,67              | 1,35                 |
| 2     | 6,66              | 4,01                 |
| 3     |                   | 1,03                 |
| 4     |                   | 0,55                 |
| 5     |                   | 0,38                 |
| 6     |                   | 0,30                 |
| 7     |                   | 0,23                 |
| 8     |                   | 0,17                 |
| 9     |                   | 0,12                 |
| 10    |                   | 0,08                 |
| 11    |                   | 0,05                 |
| 12    |                   | 0,03                 |
| 13    |                   | 0,02                 |
| 14    |                   | 0,01                 |

```python
t = [i for i in range(1,15)]
T = len(t)
P = [1.67, 6.66] + [0] * (T-2)
Q = [
    1.35, 4.01, 1.03, 0.55, 0.38, 0.30, 0.23, 
    0.17, 0.12, 0.08, 0.05, 0.03, 0.02, 0.01,
]
```

### Determinação do valor de $K$

#### A - Distância entre o CG do Histograma e do Hietograma

$$
    K = \dfrac{\sum_{i=1}^{n} Q_i t_i}{\sum_{i=1}^{n} Q_i} - \dfrac{\sum_{i=1}^{n} P_i t_i}{\sum_{i=1}^{n} P_i}
$$

```python
Qxt = [Q[i] * t[i] for i in range(T)]
Pxt = [P[i] * t[i] for i in range(T)]
```

```python
KA = (sum(Qxt) / sum(Q)) - (sum(Pxt) / sum(P))
print("KA =", round(KA, 3))
```

As ordenadas $h_i$ do hidrograma unitário obtido com $K = KA$ são dadas por:
$$
    h_i = \int_{0}^{1} \mu(i-\tau)d\tau = e^{\frac{-i}{K}} - e^{\frac{-(i+1)}{K}}
$$

```python
huA = [np.exp(-i/KA) - np.exp(-(i+1)/KA) for i in range(T)]
```

Vazão simulada:

```python
QA = [0] * T
for i in range(T):
    if i == 0:  # primeiro instante de tempo
        QA[i] = P[0] * huA[0]
    else:
        QA[i] = P[0] * huA[i] + P[1] * huA[i-1]
```

#### B - Minimização da distância quadrática entre armazenamento calculado e observado

$$
    K = \dfrac{\sum_{i=1}^{n} S_i Q_i}{\sum_{i=1}^{n} Q_i^2}
$$

Volume armazenado na bacia:

```python
S = [0] * T
for i in range(T):
    S[i] = S[i-1] + (P[i] - Q[i])
```

```python
QxQ = [Q[i] * Q[i] for i in range(T)]
SxQ = [S[i] * Q[i] for i in range(T)]
```

```python
KB = sum(SxQ) / sum(QxQ)
print("KB =", round(KB, 3))
```

As ordenadas $h_i$ do hidrograma unitário obtido com $K = KB$ são dadas por:
$$
    h_i = \int_{0}^{1} \mu(i-\tau)d\tau = e^{\frac{-i}{K}} - e^{\frac{-(i+1)}{K}}
$$

```python
huB = [np.exp(-i/KB) - np.exp(-(i+1)/KB) for i in range(T)]
```

Vazão simulada:

```python
QB = [0] * T
for i in range(T):
    if i == 0:  # primeiro instante de tempo
        QB[i] = P[0] * huB[0]
    else:
        QB[i] = P[0] * huB[i] + P[1] * huB[i-1]
```

```python
plt.plot(t, Q, marker="o", label="Vazão histórica")
plt.bar(t, P, width=0.5, label="Precipitação", alpha=0.2)
plt.plot(t, QA, marker="^", label="Vazão simulada com $K = K_A$")
plt.plot(t, QB, marker="s", label="Vazão simulada com $K = K_B$")
plt.xlabel("Tempo ($\Delta t$)")
plt.ylabel("P (mm) e Q (mm/$\Delta t$)")
plt.legend()
plt.tight_layout()
# plt.savefig("lista3_ex1.png", format="png", dpi=300)
plt.show()
```

### Maximização do coeficiente de Nash-Sutcliffe

```python
r2A = he.nse(np.array(QA), np.array(Q))
print("R2A =", r2A)
```

```python
r2B = he.nse(np.array(QB), np.array(Q))
print("R2B =", r2B)
```

```python
import numpy as np
import hydroeval as he
import matplotlib.pyplot as plt
from scipy import integrate

# Ajustando gráficos inline
plt.rcParams['figure.dpi'] = 150
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
```

Determinar o HUI de um modelo composto por dois reservatórios lineares em série nos quais a entrada da precipitação se distribui uniformemente nos dois reservatórios (adotar o mesmo valor de $K$ para os dois reservatórios);

$$
\mu(t) = \dfrac{1}{2K} e^{\frac{-t}{K}} + \dfrac{1}{2K^2} te^{\frac{-t}{K}}
$$

Comparar o seu resultado graficamente com o resultado apresentado em sala, no qual a bacia também era representada por dois reservatórios em série, mas a entrada da chuva se dava somente no primeiro reservatório. O que você conclui?


```python
K = 1
t = [i for i in range(0, 14)]
```

Reservatório Linear Simples com 2 reservatórios e P distribuído:

```python
u = lambda t, K : np.exp(-t/K) / (2 * K) + t * np.exp(-t/K) / (2 * K**2)

u_rls2 = []  # HUI para RLS com P distribuído
for ti in t:
    u_rls2.append(u(ti,K))
```

Modelo Nash para 2 reservatórios:

$$
\mu(t) = \dfrac{t}{K^2} e^{\frac{-t}{K}}
$$

```python
u_nash2 = lambda t, K : t * np.exp(-t/K) / (K**2)

u_nash = []  # HUI para modelo Nash com 2 reservatórios
for ti in t:
    u_nash.append(u_nash2(ti,K))
```

```python
plt.plot(t, u_rls2, marker="o", label="RLS com P distribuído")
plt.plot(t, u_nash, marker="^", label="Nash com 2 reservatórios")
plt.xticks([i for i in range(0,14,2)])
plt.xlabel("Tempo ($\Delta t$)")
plt.ylabel("$\mu(t)$")
plt.legend()
plt.tight_layout()
plt.savefig("rls_2.png", format="png", dpi=300)
plt.show()
```

```python
import numpy as np
import hydroeval as he
import matplotlib.pyplot as plt
from scipy import integrate

# Ajustando gráficos inline
plt.rcParams['figure.dpi'] = 150
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
```

Determinar o HUI para uma bacia que funcione como dois reservatórios em paralelo.
Adote, por hipótese, que cada reservatório recebe uma parcela $P1$ e $P2$ da chuva unitária ($P1 + P2 = 1$) e que cada reservatório possui valores de $K1$ e $K2$ distintos entre si;

$$
\mu(t) = \dfrac{P_1}{K_1} e^{\frac{-t}{K_1}} + \dfrac{P_2}{K_2} e^{\frac{-t}{K_2}}
$$

```python
u = lambda t, K1, K2, P1, P2 : P1 * np.exp(-t/K1) / K1 + P2 * np.exp(-t/K2) / K2
```

Comparar graficamente o HUI para $P1 = 0.3$ e para valores de $\frac{K2}{K1} = 1$, $2$, $5$ e $10$ (adotar $K1 = 1$). O que você conclui?

```python
P1 = 0.7
P2 = 1 - P1
K1 = 1
K = [1, 2, 5, 10]
t = [i for i in range(1, 20)]
```

```python
u_t = []
for K2 in K:
    u_t.append([])
    for ti in t:
        u_t[-1].append(u(ti, K1, K2, P1, P2))
```

```python
marker = ["o", "^", "s", "X"]
for i in range(len(K)):
    plt.plot(t, u_t[i], label="$K_2$ = "+str(K[i]), marker=marker[i])
plt.xlabel("Tempo ($\Delta t$)")
plt.ylabel("$\mu(t)$")
plt.xticks(range(1, 21, 2))
plt.legend()
plt.tight_layout()
plt.savefig("rls_3_P1modif.png", format="png", dpi=300)
plt.show()
```
